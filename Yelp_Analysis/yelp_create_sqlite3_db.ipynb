```python
import json
import sqlite3

# connect to database
db = sqlite3.connect('yelp.db')
c = db.cursor()
```

```python
# create user table
c.execute('''CREATE TABLE user
	         (user_id text,
              name text,
              review_count integer,
              average_stars float,
              fans integer,
              yelping_since date,
              type text,
              votes_funny integer,
              votes_useful integer,
              votes_cool integer)''')

# load json file
users = []
with open('../Data_Sets/yelp_phoenix_academic_dataset2/yelp_academic_dataset_user.json', 'rb') as file_obj:
    users = json.loads(file_obj.read())

# nested loops for loading data from reviews to review table
user_json_fields = ['user_id','name','review_count','average_stars','fans','yelping_since','type']
user_votes_json_fields = ['funny','useful','cool']


for i in range(len(users)):
    row_items=[]
    for j in user_json_fields:
        row_items.append(users[i][j])
    for k in user_votes_json_fields:
        row_items.append(users[i]['votes'][k])
    c.execute('INSERT INTO user VALUES (?,?,?,?,?,?,?,?,?,?)', row_items)
```

```python
# create business table
c.execute('''CREATE TABLE business
	         (business_id text,
	          name text,
              stars float,
              review_count integer,
	          full_address text, 
	          lattitude float,
              longitude float,
              open boolean,
              type text)''')

# load json file
business = []
with open('../Data_Sets/yelp_phoenix_academic_dataset2/yelp_academic_dataset_business.json', 'rb') as file_obj:
    business = json.loads(file_obj.read())

# nested loops for loading data from reviews to review table
business_json_fields = ['business_id','name','stars','review_count','full_address','latitude','longitude',
                      'open','type']

for i in range(len(business)):
    row_items=[]
    for j in business_json_fields:
        row_items.append(business[i][j])
    c.execute('INSERT INTO business VALUES (?,?,?,?,?,?,?,?,?)', row_items)
```

```python
# create reviews table
c.execute('''CREATE TABLE review
	         (review_id text,
	          user_id text,
              business_id text,
	          stars integer,
	          review_date date,
	          review_text text,
              funny integer,
              useful integer,
              cool integer)''')

# load json file
reviews = []
with open('../Data_Sets/yelp_phoenix_academic_dataset2/yelp_academic_dataset_review.json', 'rb') as file_obj:
    reviews = json.loads(file_obj.read())

# nested loops for loading data from reviews to review table
review_json_fields = ['review_id','user_id','business_id','stars','date','text']
votes_json_fields = ['funny','useful','cool']

for i in range(len(reviews)):
    row_items=[]
    for j in review_json_fields:
        row_items.append(reviews[i][j])
    for k in votes_json_fields:
        row_items.append(reviews[i]['votes'][k])
    c.execute('INSERT INTO review VALUES (?,?,?,?,?,?,?,?,?)', row_items)

```

```python
# create business categories table
c.execute('''CREATE TABLE business_categories
	         (business_id text, category text, is_true integer)''')
    
# nested loops for loading data from reviews to review table
for i in range(len(business)):
    for k in range(len(business[i]['categories'])):
        row_items=[]
        row_items.append(business[i]['business_id'])
        row_items.append(business[i]['categories'][k])
        row_items.append(True)
        c.execute('INSERT INTO business_categories VALUES (?,?,?)', row_items)
```

```python
# commit changes to table and close database
db.commit()
db.close()  
```
