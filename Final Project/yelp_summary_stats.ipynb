```python
import sqlite3
import prettytable as pt
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# connect to database
db = sqlite3.connect('yelp.db')
c = db.cursor()

# create Pandas Series for review counts by star rating
review_count = pd.Series()
review_count.name = 'review_count'

for i in ['1','2','3','4','5']:    
    c.execute("SELECT count(*) FROM review, business_categories WHERE stars = ? AND business_categories.business_id = "+
              "review.business_id AND business_categories.category = ?", (i,'Restaurants'))
    review_count[i] = c.fetchone()[0]

review_count.index.name = 'star_rating'

print "sum of reviews = %d" % sum(review_count)
```

```python
stars_review = []
for i in c.execute("SELECT stars FROM review, business_categories WHERE business_categories.business_id = review.business_id "+
                   "AND business_categories.category = 'Restaurants'"):
    stars_review.append(i[0])
    
stars_review = np.array(stars_review)
stars_sd =np.std(stars_review)
```

```python
print review_count
pd.Series.plot(review_count,kind='bar',use_index=True,title='Count of Ratings by Stars')
```

```python
total = 0
for i in review_count.index:
    total += review_count[i] * float(i)

average = total/sum(review_count)
    
print "Mean review was %f stars" % average
print "The standard deviation of reviews was %f stars" % stars_sd
```
